```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("D:/PythonProject/ANN Implementation/aps_data.csv")
```

```python
dataset = dataset.replace("na", np.nan)
```

```python
pd.DataFrame(dataset.isnull().sum().sort_values(ascending=False)).rename(columns={0:"Null Value Count"})
```

```python
### Filling null values in feature with median of that feature
for feature in [feature for feature in dataset.columns if feature not in ['class']]:
    dataset[feature] = pd.to_numeric(dataset[feature], errors='coerce')
    dataset[feature] = dataset[feature].fillna(dataset[feature].median())
```

```python
X = dataset.drop("class", axis=1)
```

```python
y = dataset["class"]
```

```python
X.info(verbose=True, show_counts=True)
```

```python
# y = pd.get_dummies(y,drop_first=True)
```

```python
label_encoder = LabelEncoder()
y = pd.DataFrame(label_encoder.fit_transform(y))
print(y)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
classifier = Sequential()

classifier.add(Dense(units=10, kernel_initializer="he_uniform", input_dim = 170))

classifier.add(Dense(units=10, kernel_initializer="he_uniform", activation="relu"))

classifier.add(Dense(units=1, kernel_initializer="glorot_uniform", activation="sigmoid"))
```

```python
classifier.compile(optimizer= 'adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
classifier.summary()
```

```python
model_history = classifier.fit(X_train, y_train, batch_size=10, epochs=20, validation_split=0.25)
```

```python
model_history.history.keys()
```

```python
model_history.history["accuracy"]
```

```python
plt.plot(model_history.history["accuracy"])
plt.plot(model_history.history["val_accuracy"])
plt.title('Model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(model_history.history["loss"])
plt.plot(model_history.history["val_loss"])
plt.title('Model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred = y_pred>0.05
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred=y_pred)
```

```python
from keras_tuner.tuners import RandomSearch
```

```python
def build_model(hp):
    model = Sequential()
    model.add(Dense(32, kernel_initializer='he_uniform',activation='relu', input_dim = 170))
    model.add(Dense(32, kernel_initializer='he_uniform', activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    optimizer = hp.Choice("optimizer", values= ["adam", "sgd", "rmsprop", "adadelta"])
    model.compile(optimizer=optimizer, loss = 'binary_crossentropy', metrics=['accuracy'])
    return model

    
```

```python
tuner = RandomSearch(build_model, objective='val_accuracy', max_trials=5, directory='project', project_name = 'aps_fault')
```

```python
tuner.search(X_train, y_train, validation_data=(X_test, y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
my_model = tuner.get_best_models(num_models=1)[0]
```

```python
my_model.summary()
```

```python
from tensorflow.keras import optimizers
def build_model_1(hp):
    model = Sequential()
    for i in range(hp.Int('num_layers',2,20)):
        model.add(Dense(units=hp.Int('unit_' + str(i), min_value=32, max_value=512,
                                     step=32),
                                     activation='relu'))
    model.add(Dense(1, activation="sigmoid"))
    model.compile(optimizers.Adam(
            hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),
            loss = 'binary_crossentropy',
            metrics=['accuracy'])
    return model
```

```python
tuner_second = RandomSearch(build_model_1, objective='val_accuracy', max_trials=5, directory = 'project1', project_name = 'aps_fault')
```

```python
tuner_second.search(X_train, y_train, epochs=5, validation_data=(X_test,y_test))
```

```python
tuner_second.get_best_hyperparameters()[0].values
```

```python
my_model_1 = tuner_second.get_best_models(num_models=1)[0]
```

```python
my_model_1.summary()
```

```python
import time
import os

def saveModel_path(model_dir="SAVED_MODELS"):
    os.makedirs(model_dir,exist_ok=True)
    fileName = time.strftime("Model_%Y_%m_%H_%M_%S_.h5")
    model_path = os.path.join(model_dir, fileName)
    print(f"your model will be saved at the following location\n{model_path}")
    return model_path
```

```python
my_model_1.save(saveModel_path())
```
